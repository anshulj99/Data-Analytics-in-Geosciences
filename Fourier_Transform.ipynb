```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#constructing a time signal
Fs = 2000 #Sampling frequency of 2KHz
time_step = 1/Fs #sample time interval

f0 = 100 #Signal frequency of 100Hz

N= int(Fs/f0) #Number of sample of signals

t = np.linspace(0, (N-1)*time_step, N)

f_step = Fs/N #Frequency interval
f = np.linspace(0, (N-1)*f_step, N )

y = 1.1 * np.sin(2 * np.pi * f0 * t) #signal with 1.1 amplitude
```

```python
#plot
fig, axs = plt.subplots(1,1)
axs.plot(t,y,'.-')
axs.set_ylabel("Amplitude")
axs.set_xlabel("Time")
```

```python
#performing fft
yf = np.fft.fft(y) #has complex values
yf_mag = np.abs(yf)
yf_normalized = yf_mag/N
```

```python
#plot
fig, axs = plt.subplots(2,1)
axs[0].plot(t,y,'.-',color ='green')
axs[1].plot(f, yf_normalized,color='red')
axs[0].set_ylabel("Amplitude")
axs[0].set_xlabel("Time")
axs[1].set_ylabel("Amplitude")
axs[1].set_xlabel("Frequency")
plt.subplots_adjust(hspace=0.7)
```

```python
#using sampling theorem, if we have frequency of 2kHz, we will be able to construct a signal at Fs/2
f_plot = f[0:int(N/2)]
yf_plot = 2 * yf_normalized[0:int(N/2)]
```

```python
fig, axs = plt.subplots(2,1)
axs[0].plot(t,y,'.-',color ='red')
axs[1].plot(f_plot, yf_plot,color ='green')
axs[0].set_ylabel("Amplitude")
axs[0].set_xlabel("Time")
axs[1].set_ylabel("Amplitude")
axs[1].set_xlabel("Frequency")
plt.subplots_adjust(hspace=0.7)
```
